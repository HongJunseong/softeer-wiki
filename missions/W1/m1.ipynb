## head, tail, shape

```python
import pandas as pd

# data
data = pd.read_csv('mtcars.csv')

# head
display(data.head())

# tail
display(data.tail())

# shape
print(data.shape)
```

## Rename Column

```python
# noname columns
unnamed_cols = data.columns[data.columns.str.contains('Unnamed')]
data.rename(columns={unnamed_cols[0]: 'car_name'}, inplace=True)

display(data.head())
```

## info, dtypes

```python
# info
data.info()

# dtypes
data.dtypes
```

## describe

```python
# describe
data.describe()
```

## Count unique values

```python
# count unique values (gears, transmission)
print(data["gear"].nunique()) # gears   
print(data["am"].nunique())   # transmission
```

## Visualization

```python
import matplotlib.pyplot as plt

# combination of gear and transmission
combination = data.groupby(['gear', 'am']).size()
print(combination)

# bar plot
combination.plot(kind='bar', figsize=(10, 5))

plt.title("# of Cars by Gear x Transmission")
plt.ylabel("# of Cars")
plt.xlabel("(Gear, Auto Transmission)")
plt.show()
```

```python
# crosstab

import matplotlib.pyplot as plt

table = pd.crosstab(data['gear'], data['am'])
print(table)

table.plot(kind='bar')

plt.title("# of Cars by Gear x Transmission")
plt.ylabel("# of Cars")
plt.xlabel("(Gear, Auto Transmission)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# histograms
data.hist(figsize=(15, 10))

plt.suptitle("Histograms of Variables")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

# scatter plot
plt.scatter(data['cyl'], data['hp'])

plt.title("Cylinder vs Horse Power")
plt.xlabel("Cylinder")
plt.ylabel("Horse Power")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

# scatter plot
plt.scatter(data['mpg'], data['hp'])

plt.title("MPG vs Horse Power")
plt.xlabel("MPG")
plt.ylabel("Horse Power")
plt.show()
```

## Correlation

```python
# correlation matrix
correlation_matrix = data.select_dtypes(include="number").corr()
display(correlation_matrix)
```
